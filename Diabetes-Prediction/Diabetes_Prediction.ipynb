```python
import numpy as np #Numpy arrays
import pandas as pd #creating data frames in a structured table
from sklearn.preprocessing import StandardScaler #used to stardize data into a normal range
from sklearn.model_selection import train_test_split#used to split data into training and testing
from sklearn import svm #used to import SUpportVectorMachine algorithm
from sklearn.metrics import accuracy_score#used to find the accuracy of the final test
```

Data collection and Analysis
PIMA diabetes dataset

```python
#loading the dataset to a pandas file
diabetes_dataset = pd.read_csv('/content/diabetes.csv')
```

```python
#the head function helps to bring the first 5 rows of the dataset
diabetes_dataset.head()
```

```python
#number of rows and column, shape helps you to know the dimension of the data
diabetes_dataset.shape
```

```python
# Statistical measures of the data
diabetes_dataset.describe()
```

```python
diabetes_dataset['Outcome'].value_counts()
```

0----> Non diabetic


1----> Diabetic


```python
diabetes_dataset.groupby('Outcome').mean()
```

```python
#seperating the data and labels
X = diabetes_dataset.drop(columns = 'Outcome', axis=1)
Y = diabetes_dataset['Outcome']
```

Data Standardization

```python
scaler=StandardScaler()
```

```python
scaler.fit(X)
```

```python
standardized_data = scaler.transform(X)
```

```python
X= standardized_data
Y= diabetes_dataset['Outcome']
```

Train test Split

```python
nan_indices = Y.isna()
X_cleaned = X[~nan_indices]
Y_cleaned = Y[~nan_indices]

X_train, X_test, Y_train, Y_test = train_test_split(X_cleaned, Y_cleaned, test_size = 0.2, stratify=Y_cleaned, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Training the model

```python
classifier=svm.SVC(kernel='linear')
```

```python
#training the support vector machine classifier
classifier.fit(X_train,Y_train)
```

MODEL EVALUATION

Accuracy Score

```python
#accuracy score on the training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print(training_data_accuracy)
```

```python
#accuracy score on the test data
#accuracy score on the training data
X_train_prediction = classifier.predict(X_test)
training_data_accuracy = accuracy_score(X_train_prediction, Y_test)
```

```python
print(training_data_accuracy)
```

Making a Predictive System

```python
input_data=(1,5,166,72,19,175,25.8,0.587,51)
#change the input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

#reshape the array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)
#reshpe the array for predicting one instance
std_data = scaler.transform(input_data_reshaped)
prediction = classifier.predict(std_data)
print(prediction)
```
